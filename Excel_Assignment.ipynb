  1. Explain the difference between Relative, Absolute and Mixed Cell Referencing.

 - These three types of cell referencing are fundamental to working with formulas in spreadsheets (like Microsoft Excel or Google Sheets) as they determine how cell addresses are adjusted when a formula is copied to other cells.

 Relative Cell Referencing (e.g., A1):
 Behavior: The cell address changes relatively when the formula is copied to another cell. If you copy a formula one column right and one row down, the cell reference will also shift one column right and one row down.

 Use Case: Ideal for performing an operation across a whole range of data, like  multiplying corresponding cells in two columns.

 Absolute Cell Referencing (e.g., $A$1):
 Behavior: The cell address remains fixed (or "absolute") no matter where the formula is copied. The dollar sign ($) locks both the column and the row.

 Use Case: Essential when a formula needs to consistently refer back to a single, specific cell (e.g., a fixed tax rate or a conversion factor) even when copied down or across multiple rows/columns.

 Mixed Cell Referencing (e.g., $A1 or A$1):
 Behavior: One part of the cell address is absolute (locked with a $), and the other part is relative (unlocked).

 $A1 locks the column (A), but the row (1) will change.

 A$1 locks the row (1), but the column (A) will change.

 Use Case: Useful for creating things like multiplication tables or when you need to copy a formula across a range but only want to lock the reference to a specific row header or column header.
